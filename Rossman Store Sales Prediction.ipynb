## Visualization

```python
# Visualize for every day of the week the sales and the customers scatterplot

plt.rcParams['figure.figsize'] = (15.0, 12.0)

f, ax = plt.subplots(7, sharex = True, sharey = True)
for i in range (1, 8):
    data = train[train['DayOfWeek'] == i]
    ax[i - 1].set_title('Day {0}'.format(i))
    ax[i - 1].scatter(data['Customers'], data['Sales'], label = i)
    
plt.legend()
plt.xlabel('Customers')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()    
```

```python
# Combined scatterplot for customer, sales and days

plt.scatter(train['Customers'], train['Sales'], c = train['DayOfWeek'], alpha = 0.9, cmap = plt.cm.get_cmap('YlGn'))

plt.xlabel('Customers')
plt.ylabel('Sales')
plt.show()
```

```python
# Change categorical values to categories

store['StoreType'] = store['StoreType'].astype('category').cat.codes
store['Assortment'] = store['Assortment'].astype('category').cat.codes
train['StateHoliday'] = train['StateHoliday'].astype('category').cat.codes
```

```python
# Merge store and train dataset

merged = pd.merge(train, store, on = 'Store', how = 'left')
merged.head()
```

```python
merged.isnull().sum()
```

```python
# Remove all the null values with zero

merged.fillna(0, inplace = True)
```

```python
# 'Date' variable is still Object datatype. We will change it to date.

merged['Date'] = pd.to_datetime(merged['Date'])
merged.dtypes
```

```python
# We need to split 'Date' into 'Year', 'Month', 'Day', 'Week' to use it in our ML model

merged['Year'] = merged.Date.dt.year
merged['Month'] = merged.Date.dt.month
merged['Week'] = merged.Date.dt.week
merged['Day'] = merged.Date.dt.day
merged.head()
```

```python
# Create a column with the competition open in months
merged['MonthsCompetitionOpen'] = 12 * (merged['Year'] - merged['CompetitionOpenSinceYear']) + (merged['Month'] - merged['CompetitionOpenSinceMonth'])
merged.loc[merged['CompetitionOpenSinceYear'] == 0, 'MonthsCompetitionOpen'] = 0
```

```python
merged['WeeksPromoOpen'] = 12 * (merged['Year'] - merged['Promo2SinceYear']) + (merged['Date'].dt.isocalendar().week - merged['Promo2SinceWeek'])
merged.loc[merged['Promo2SinceYear'] == 0, 'WeekPromoOpen'] = 0
```

```python
merged.dtypes
```

```python
# We transform all floats into int because it will be easier to handle

toInt = [
    'CompetitionOpenSinceMonth',
    'CompetitionOpenSinceYear',
    'Promo2SinceWeek',
    'Promo2SinceYear',
    'WeeksPromoOpen',
    'MonthsCompetitionOpen']

merged[toInt] = merged[toInt].astype(int)
```

```python
# Get Median sales to understand the distrubution

med_store = train.groupby('Store')[['Sales','Customers','SalesPerCustomer']].median()
med_store.rename(columns=lambda x: 'Med' + x, inplace = True)

store = pd.merge(med_store.reset_index(), store, on = 'Store')
```

## Model Building and Evaluation

```python
from sklearn.model_selection import train_test_split

X = [
    'Store',
    'Customers',
    'CompetitionDistance'
    
    'StoreType', 
    'Assortment', 
    'CompetitionDistance',
    
    'DayOfWeek', 
    'Year', 
    'Month',
    'Week', 
    'Day',
   
    'Promo',
    'Promo2',
    
    'AvgSales'
    'AvgCustomers'
    'AvgSalesPerCustomer',
    
    'StateHoliday',
    'SchoolHoliday',  
        
    'CompetitionOpenSinceMonth', 
    'CompetitionOpenSinceYear', 
    'Promo2SinceWeek', 
    'Promo2SinceYear',
    
    'MedSales',
    'MedCustomers',
    'MedSalesPerCustomer']

X_data = merged[X]
Y_data = np.log(merged['Sales'])   # our sales histogram is skewd so we have to log transform the data\
X_train, X_test, y_train, y_test = train_test_split(X_data, Y_data, test_size = 0.20, random_state = 10)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error

def plot_importance(model):
    k = list(zip(X, model.feature_importances_))
    k.sort(key = lambda tup: tup[1])
    
    lables, vals = zip(*k)
    
    plt.barh(np.arrange(len(X)), vals, align = 'center')
    plt.yticks(np.arrange(len(X)),labels)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV

param = {
        'n_estimators': [100,500,1000,1500],
        'max_depth': [2,4,6,8]
        }

xgboost_tree = xgb.XGBRegressor(
    eta = 0.1,
    min_child_weight = 2,
    subsample = 0.8,
    colsample_bytree = 0.8,
    tree_method = 'exact',
    reg_alpha = 0.05,
    silent = 0,
    random_state = 1023
)

grid = GridSearchCV(estimator=xgboost_tree, param_grid=param, cv=5, verbose=1, n_job=-1, scoring = 'neg_mean_squared_error')

grid_result = grid.fit(X_train, y_train)
best_params = grid_result.best_params_

print('Best Params:', best_params)
```

```python
from math import sqrt

pred = grid_result.predict(X_test)
print('Root Mean squared error {}'.format(sqrt(mean_squared_error(np.exp(y_test), np.exp(pred)))))
```
